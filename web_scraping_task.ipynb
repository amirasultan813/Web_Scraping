```python
import scrapy
from scrapy.crawler import CrawlerProcess
import pandas as pd
```

```python
class DC_Chapter_Spider(scrapy.Spider):
    name='dc_chapter_spider'
    def start_requests(self):
        url='https://eg1lib.org/s/social%20media'
        yield scrapy.Request(url=url,callback=self.parse_front)
    def parse_front(self,response):
        book_names=response.css('h3>a::text')
        writer_names=response.xpath('//div[@class="authors"]/a[1][@class="color1"]/text()').extract()
        year=response.xpath('//div[@class="bookDetailsBox"]/div[@class="bookProperty property_year"]/div[2]/text()')
        ch_books=[t.strip() for t in book_names.extract()]
        ch_writer=[t.strip() for t in writer_names]
        ch_year=[t.strip() for t in year.extract()]
        dc_dict['Book_Names']=ch_books
        dc_dict['Writer_Names']=ch_writer
        dc_dict['Year_Published']=ch_year
        #print(book_names)
        print(len(writer_names))
        print(len(book_names))
        print(len(year))
dc_dict=dict()
process=CrawlerProcess()
process.crawl(DC_Chapter_Spider)
process.start()
```
